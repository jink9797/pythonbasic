```python
import time

results = []
url = "https://openapi.kric.go.kr/openapi/safetyInfo/stationDefibrillator"
params = {"serviceKey":"$2a$10$pXB1DRp8b0rER7lkbUNHtOZCdCTkZnmSkp3cbrmSQ3tOaBhYgB1Rq",
          "format":"JSON",
          "lnCd":"",
          "railOprIsttCd":"",
          "stinCd":""}

first_time = time.time()
for idx in range(len(codes)):
    _start_time = time.time()
    
    # params setting
    params["lnCd"] = codes.at[idx, "RAIL_OPR_ISTT_CD"]
    params["railOprIsttCd"] = codes.at[idx, "RAIL_OPR_ISTT_CD"]
    params["stinCd"] = codes.at[idx, "RAIL_OPR_ISTT_CD"]
    
    # url로 params를 가지고 HTTP 통신을 요청하고 그 결과를 받아옵니다.
    response = requests.get(url, params=params)
    
    # 통신이 성공적으로 수행되었다면
    if response.status_code == 200:
        # HTTP 통신으로 받아온 response body 값을 dict 형태로 변환합니다. 
        json_dict = json.loads(response.text)
        
        # dict 객체를 results 리스트 자료형에 담습니다.
        results.append(json_dict)
        
        # 안내 문구를 출력합니다.
        print(f'{idx}번째 데이터를 처리했습니다. 소요시간: {round(time.time() - _start_time, 2)}s')
print(f'총 소요시간: {round(time.time() - first_time)}s')
```
