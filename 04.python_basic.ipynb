```python
### for문과 range() 내장함수
#  - range()내장함수는 연속된 숫자를 만드는데 특화되어 있다.
  # - for문은 반복 실행하는데 특화되어 있다.
  #   이 둘을 활용하여 print('hello world') 1,000번 출력됨 

for n in range(1000):
    print("Hello world")

## for 요소를 담을 변수 in 반복 가능한 객체:
##  반복할 코드(띄어쓰기 4칸)

## for 변수 in range(횟수):
##  반복할 코드(띄어쓰기 4칸)
```

```python
## range()함수는 슬라이싱과 구조가 비슷하다.

## range변수 = range(시작,끝,증감크기)
## 시퀀스자료형변수 [시작:끝:증감크기]

# range()함수로 list 생성하기 / unpacking도 가능함

# ex) 100미만의 2의 배수를 담은 리스트를 생성하는 방법입니다.
even_numbers = list(range(2,100,2))
print(even_numbers)
 
# *** range()함수 특징 ***
# - 돌려준 객체는 마치 리스트인것처럼 동작하지만, 사실 리스트와는 다르다.
# - range 객체는 iterate할 때 원하는 시퀀스 항목들을 순서대로 요소를 돌려주는 객체이지만, 메모리에 공간을 할당하여 미리 모든 요소들을 만들어주는 것이
# - 아니라 필요할때 그때그때 다음 요소를 만들어 반환하기에 공간 효율적이다.
```

```python
# Q 계승(factorial)을 계산, 사용자로부터 정수를 하나 입력받습니다.
     # 입력받은 숫자까지의 계승을 for문을 활용 해 구한 뒤 출력
     # - 계승이란 1부터  n까지의 자연수를 모두 곱하는 것을 의미
     # >>>> n! = 1 * 2 * 3 * ..... * (n - 1) * n
```

```python
fac = int(input("실행결과:\n계승을 구할 숫자를 입력하세요."))     #range 는 시작점은 포함 / 끝은 포함안함
f = 1
for n in range(1, fac + 1):        # 1~3 /+1  
    f = f * n
print(f)
```
