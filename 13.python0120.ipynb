### DataFrame apply()메서드
- 메서드는 첫 인자로 함수를 필수 값으로 받음(경우에 따라 두번째 인자로 axis를 사용)
- axis 가 0 or index인 경우 각 column에 대해 함수 적용
- axis 가 1 or coluns인 경우 각 row에 대해 함수 적용
#### DataFrame.apply(func, axis = 0, row = False, result_type = None, args = (), **kargs)

- np.sqrt 는 각 요소마다 적용
- np.sqrt(df)와 동일한

```python
import pandas as pd
import numpy as np
df = pd.DataFrame([[4,9]]*3, columns = ['A','B'])
df
```

### 실수 값을 카테고리 값으로 변환
실수 값을 크기 기준 카테고리 값으로 변환하고 싶을때
- cut : 실수 값의 경계선을 지정하는 경우 
x= 1차원 형태의 배열 형태가 옵니다
bins = int, 스칼라를 요소로 갖는 시퀀스가 옵니다.
- qcut : 개수가 똑같은 구간으로 나누는 경우 (분위수)
x = 1d ndarray 혹은 시리즈
q = in 혹은 분위수를 나타내는 1 이하의 실수를 요소로 갖는 list (e.g.[0, .25, .5, .75, 1.])

```python
ages = [0,2,10,21,23,37,31,61,20,41,32,101]
```

#### cut 명령을 사용 -> 실수값을 다음처럼 카테고리 값으로 바꿀 수 있다.
#### bins 인수는 카테고리를 나누는 기준값이 된다. 영역을 넘는 값은 NaN으로 처리

```python
bins = [1,20,30,50,70,100]
labels = ['미성년자','청년','장년','중년','노년']
cats = pd.cut(ages, bins, labels = labels)
cats
```

```python
type(cats)
```

```python
cats.categories
```

```python
cats.codes
```

#### cut() 명령이 반환하는 값 : Categorical 클래스 객체

```python
df4 = pd.DataFrame(ages,columns = ['ages'])
df4['age_cat'] = pd.cut(df4.ages, bins, labels=labels)
df4
```

### 합성
pandas는 두 개 이상의 DataFrame을 하나로 합치는 데이터 병합(merge)나 연결(concatenate)을 지원

DataFrame merge()
- merge()함수는 두 데이터 프레임의 공통 column
- 인덱스를 기준으로 두개의 테이블을 합침
- 기준이 되는 column, row의 데이터를 키(key)라고 함

```python
df1 = pd.DataFrame({
    '고객번호': [1001, 1002, 1003, 1004, 1005, 1006, 1007],
    '이름': ['둘리', '도우너', '또치', '길동', '희동', '마이콜', '영희']
}, columns=['고객번호', '이름'])
df1
```

```python
f2 = pd.DataFrame({
    '고객번호': [1001, 1001, 1005, 1006, 1008, 1001],
    '금액': [10000, 20000, 15000, 5000, 100000, 30000]
}, columns=['고객번호', '금액'])
df2
```

### DataFrame merge()
merge 함수로 위의 두 DataFrame df1, df2를 합치면 공통 column인 고객번호 column을 기준으로 데이터를 찾아서 합침
기본적으로 양쪽 DataFrame에 모두 키가 존재하는 데이터만 보여주는 inner join방식을 사용

```python
pd.merge(df1,df2)
```

outer join 방식은 키 값이 한쪽에만 있어도 데이터를 보여줌

```python
pd.merge(df1, df2, how = 'outer')
```

left는 첫번째 인수 기준
right는 두번째 인수 기준으로 DataFrame의 키값을 모두 보여줌

```python
pd.merge(df1,df2, how = 'left')
```

```python
pd.merge(df1, df2, how = 'right')
```
