### Axes
- 숫자 인덱싱은 파이썬이 아닌Matlab관행을 따르기 때문에 
- ----> 첫번째 플롯을 가르키는 숫자가 0이 아니라 1임에 주의
- tight_layout()명령을 실행하면 플롯 간의 간격 자동으로 맞춰짐

```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2* np.pi *x1) *np.exp(-x1)
y2 = np.cos(2* np.pi *x2)

ax1 = plt.subplot(2,1,1)
plt.plot(x1,y1, 'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

ax2 = plt.subplot(2,1,2)
plt.plot(x2,y2,'r.-')
plt.xlabel('time(s)')
plt.ylabel('Undamped')

plt.tight_layout()
plt.show()
```

### 바 차트
- x 데이터가 카테고리 값인 경우에는 bar 명령과 barh 명령으로 바 차트(bar chart)시각화 가능 
- 가로 방향으로 바 차트를 그리려면 barh 명령을 사용
- 바 차트 작성시 주의점은 
- ----> 첫번째 인수인 left 가 x축에서 바(bar)의 왼쪽 변의 위치를 나타냄

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

y = [2,3,1]
x = ['가','나','다']

plt.title('Bar Chart')
plt.bar(x,y)
plt.xlabel('가나다')
plt.ylabel('빈도 수')
plt.show()
```

- 각 bar 마다 색상을 적용 가능
- 적용하고 싶은 색상을 데이터 순으로 리스트 생성
- 생성한 리스트를 bar()의 color 키워드 인수로 전달

```python
x = ['가','나','다']
y = [2,3,1]
colors = ['red','green','blue']

plt.title('Bar Chart')
plt.bar(x,y, color = colors)
plt.xlabel('가나다')
plt.ylabel('빈도 수')
plt.show()
```

- xerr 인수나 yerr 인수를 지정하면 에러 바(error bar)를 추가 가능
- alpha는 투명도를 지정. 0이면 완전 투명, 1이면 완전 불투명

```python
np.random.seed(0)

people = ['몽룡','춘향','방자','향단']
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

plt.title('Bar Chart')
plt.barh(people, performance, xerr = error, alpha = 0.4)
plt.xlabel('x 라벨')
plt.show()
```

- cmap 키워드 인수 사용

```python
fig, axes = plt.subplots(1, 4, figsize=(12, 3),
                         subplot_kw={'xticks': [], 'yticks': []})
axes[0].set_title("plt.cm.Blues")
axes[0].imshow(X, interpolation='nearest', cmap=plt.cm.Blues)
axes[1].set_title("plt.cm.Blues_r")
axes[1].imshow(X, interpolation='nearest', cmap=plt.cm.Blues_r)
axes[2].set_title("plt.BrBG")
axes[2].imshow(X, interpolation='nearest', cmap='BrBG')
axes[3].set_title("plt.BrBG_r")
axes[3].imshow(X, interpolation='nearest', cmap='BrBG_r')
plt.show()
```

- imshow 명령은 자료의 시각화를 돕기위해 다양한 2차원 인터폴레이션을 지원

```python
methods = [
    None, 'none', 'nearest', 'bilinear', 'bicubic', 'spline16',
    'spline36', 'hanning', 'hamming', 'hermite', 'kaiser', 'quadric',
    'catrom', 'gaussian', 'bessel', 'mitchell', 'sinc', 'lanczos'
]
fig, axes = plt.subplots(3, 6, figsize=(12, 6),
                         subplot_kw={'xticks': [], 'yticks': []})
for ax, interp_method in zip(axes.flat, methods):
    ax.imshow(X, cmap=plt.cm.bone_r, interpolation=interp_method)
    ax.set_title(interp_method)
plt.show()
```

### 컨투어 플롯
- 입력 변수가 x, y 두 개이고 출력 변수가 z 하나인 경우 :  3차원 자료가 됨. 
- (3차원)시각화하는 방법은 등고선(contour)을 사용하는 방법이 있음
- contour 혹은 contourf 명령을 사용
- contour는 등고선만 표시하고 contourf는 색깔로 표시
- 입력 변수 x, y는 그대로 사용할 수 없고 meshgrid 명령으로 그리드 포인트 행렬을 만듬

```python
def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 - y ** 2)


n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
XX, YY = np.meshgrid(x, y)
ZZ = f(XX, YY)

plt.title("Contour plots")
plt.contourf(XX, YY, ZZ, alpha=.75, cmap='jet')
plt.contour(XX, YY, ZZ, colors='black')
plt.show()
```

### 3D 서피스 플롯
- 3차원 플롯은 등고선 플롯과 달리 Axes3D라는 3차원 전용 axes를 생성 ---> 입체적으로 표시
- plot_wireframe, plot_surface 명령을 사용

```python
from mpl_toolkits.mplot3d import Axes3D
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
XX, YY = np.meshgrid(X, Y)
RR = np.sqrt(XX**2 + YY**2)
ZZ = np.sin(RR)

fig = plt.figure()
ax = Axes3D(fig)
ax.set_title('3D surface Plot')
ax.plot_surface(XX, YY,ZZ, rstride = 1, cstride = 1, cmap = 'hot')
plt.show()
```

### histplot() 다양한 표현법 poly

### histplot() 다양한 표현법 poly

```python
sns.histplot(
    penguins, x = 'bill_depth_mm', y = 'species', hue = 'species',legend = False)
```

### 히스토그램 - displot()
- displot() : 1차원 or 2차원 데이터의 분포를 히스토그램으로 보여줌
- rug 와 kde에 대해 동시에 표현 가능
- 표현 범위 더 넓고 많이 사용

histplot()과 동일한 

```python
sns.displot(data=penguins, x = 'flipper_length_mm')
```

### displot() kde로 표현
- kind 키워드 인수를 활용
- 기본적으로 표현되는 히스토그램의 종류(막대에서)를 kde(선으로)로 변경

```python
sns.displot(data=penguins, x = 'flipper_length_mm',kind='kde')
```

### displot() ecdf로 표현
- kind키워드 인수에 ecdf를 활용 -> 누적 분포 그래프로 표현
- empirical cumulative distribution functions : ecdf

```python
sns.displot(data=penguins, x = 'flipper_length_mm',kind='ecdf')
```

### displot() 히스토그램과 kde 동시에 표현
- kde 키워드 인수 값을 True로 전달
- 히스토그램과 kde를 동시에 표현
- kind 와 kde의 차이점을 비교하면서 생각해보기

```python
sns.displot(data=penguins, x = 'flipper_length_mm',kde = True)
```

### displot() x, y 모두 값 할당
- histplot() 동일하게 x축, y축 각각 값을 할당 -> 히트맵 형태로 그래프를 출력

```python
sns.displot(data=penguins, x = 'flipper_length_mm', y = 'bill_length_mm', hue = 'species')
```

### desplot() 2개 값을 kde로 표현
- x, y축 할당한 앞의 히스토그램을 kde로도 표현 가능
- rug = True 키워드 인수를 전달 -> rug에 대한 표현도 동시에 할 수 있음

```python
g = sns.displot(data=penguins, x = 'flipper_length_mm', y = 'bill_length_mm', kind ='kde',rug = True)
```
